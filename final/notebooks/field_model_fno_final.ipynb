# Final Field Model (Geometry-Aware FNO)

This notebook trains the new field model pipeline from `final/` without touching legacy files.

```python
from pathlib import Path
import json

from final.io import load_config
from final.dataset_builder import build_field_datasets
from final.train import train_field_model

ROOT = Path('.')
PIPE_CFG_PATH = ROOT / 'final/configs/pipeline_config.yaml'
TRAIN_CFG_PATH = ROOT / 'final/configs/train_field_fno.yaml'
print(PIPE_CFG_PATH)
print(TRAIN_CFG_PATH)
```

```python
# Step 1: Build field datasets (presets A-F)
pipe_cfg = load_config(PIPE_CFG_PATH)
summary = build_field_datasets(pipe_cfg, root=ROOT)
print(json.dumps(summary, indent=2))
```

```python
# Step 2: Train FNO on one preset (set `preset_root` in train config)
train_cfg = load_config(TRAIN_CFG_PATH)
train_summary = train_field_model(train_cfg)
print(json.dumps(train_summary, indent=2))
```

```python
# Optional: print saved summary file
out_dir = Path(train_cfg.get('output_dir'))
print((out_dir / 'summary.json').read_text())
```
